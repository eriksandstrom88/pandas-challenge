SET UP OUR DATA FOR ANALYSIS

```python
#Import Dependencies
import pandas as pd
import os
```

```python
# Open the necessary csv files as data frames by creating file paths and using pandas csv reader
student_filepath = os.path.join("students_complete.csv")
schools_filepath = os.path.join("schools_complete.csv")
```

```python
# Read the files with pandas
students_df = pd.read_csv(student_filepath)
schools_df = pd.read_csv(schools_filepath)
```

```python
# merge the two data frames with an inner join on school_name
master_df = pd.merge(schools_df, students_df, on="school_name")
master_df.head(100)
```

DISTRICT SUMMARY

```python
# make sure the data is clean, check the data types
#####master_df.count() ### un-comment
#master_df.dtypes

# calculate all the high level district summary statistics

# count the uniqe schools
school_column = master_df['school_name']
unique_schools_list = []
for name in school_column:
    if name not in unique_schools_list:
        unique_schools_list.append(name)
unique_schools = len(unique_schools_list)
# count the unique students
unique_students = len(master_df['Student ID'].unique()) # AS NUMBER WITH COMMAS
# calculate the total district budget
school_groupby_budget = master_df.groupby(['school_name']).mean('budget')
total_budget = school_groupby_budget.budget.sum() #AS MONEY

# calculate the average math score
avg_math_score = master_df['math_score'].mean() #AS % WITH 6 DECIMALS

# calculate the average reading score
avg_reading_score = master_df['reading_score'].mean() #AS % WITH 5 DECIMALS

# filter to the students passing math and count
filtered_passing_math = master_df.loc[master_df.loc[:,'math_score']>=70,:]
passing_math_scores = filtered_passing_math.loc[:,'math_score']
num_passing_math = passing_math_scores.count()
pct_passing_math = num_passing_math/unique_students*100 #AS % WITH 6 DECIMALS

#filter to students passing reading and count
filtered_passing_reading = master_df.loc[master_df.loc[:,'reading_score']>=70,:]
passing_reading_scores = filtered_passing_reading.loc[:,'reading_score']
num_passing_reading = passing_reading_scores.count()
pct_passing_reading = num_passing_reading/unique_students*100 #AS % WITH 6 DECIMALS

#filter to students passing both and count
filtered_passing_both = filtered_passing_math.loc[filtered_passing_math.loc[:,'reading_score']>=70,:]
students_passing_both = filtered_passing_both.loc[:,'Student ID']
num_passing_both = students_passing_both.count()
pct_passing_both = num_passing_both/unique_students*100 #AS % WITH 6 DECIMALS

#create a dictionary to then populate the district summary dataframe
district_summary_data = {
    'Total Schools':[unique_schools],
    'Total Students':[unique_students],
    'Total Budget':[total_budget],
    'Average Math Score':[avg_math_score],
    'Average Reading Score':[avg_reading_score],
    '% Passing Math':[pct_passing_math],
    '% Passing Reading':[pct_passing_reading],
    '% Overall Passing':[pct_passing_both]
}
district_summary_df = pd.DataFrame(district_summary_data)
district_summary_df

#KEEP WORKING FOR NOW, BUT NEED INFO ON HOW TO CHANGE NUMBER FORMAT
```

SCHOOL SUMMARY

```python
# get summary statistics by school
grouped_schools_df = master_df.groupby(['school_name'])
grouped_schools_stats_df = grouped_schools_df.mean()
grouped_schools_stats_df
```

```python
#count passing math scores by school
pass_math_only = master_df.loc[master_df.loc[:,'math_score']>=70,['school_name','math_score', 'reading_score']]
math_pass_school = pass_math_only.groupby(['school_name'])
passing_math_by_school = math_pass_school.count()
#count passing reading scores by school
pass_reading_only = master_df.loc[master_df.loc[:,'reading_score']>=70,['school_name','reading_score']]
reading_pass_school = pass_reading_only.groupby(['school_name'])
passing_reading_by_school = reading_pass_school.count()
#count passing both by school
pass_both = pass_math_only.loc[master_df.loc[:,'reading_score']>=70,['school_name','math_score','reading_score']]
pass_both_school = pass_both.groupby(['school_name'])
passing_both_by_school = pass_both_school.count()
passing_both_by_school
#merge passing reading count and passing math count
merged_m_r_counts = pd.merge(passing_math_by_school, passing_reading_by_school, on='school_name')
#merge the passing math and reading counts with the passing both count, delete columns, rename columns
merged_m_r_b_counts = pd.merge(merged_m_r_counts, passing_both_by_school, on='school_name')
passing_counts = merged_m_r_b_counts.rename(columns={'math_score_x':'Passing Math','reading_score_y':'Passing Reading','math_score_y':'Passing Both'})
passing_counts = passing_counts.drop(['reading_score_x','reading_score'],1)
passing_counts
#merge passing counts with summary data, rename columns, and srop the unneeded Student ID column
summary_draft = pd.merge(grouped_schools_stats_df,passing_counts, on='school_name')
summary_draft = summary_draft.rename(columns={'size':'Total Students','budget':'Total School Budget','reading_score':'Average Reading Score',
                                             'math_score':'Average Math Score'})
summary_draft = summary_draft.drop('Student ID',1)
#calculate the final summary data to add to the summary table
budget_per_student = summary_draft['Total School Budget']/summary_draft['Total Students']
pass_math_by_school = summary_draft['Passing Math']/summary_draft['Total Students']*100
pass_reading_by_school = summary_draft['Passing Reading']/summary_draft['Total Students']*100
pass_both_per_school = summary_draft['Passing Both']/summary_draft['Total Students']*100
#create a dataframe with the calculated summary data
school_summary_data = {
    'Per Student Budget':budget_per_student,
    '% Passing Math':pass_math_by_school,
    '% Passing Reading': pass_reading_by_school,
    '% Overall Passing':pass_both_per_school
}
#merge the calculated summary data with the existing summary dataframe
summary_data_by_school = pd.DataFrame(school_summary_data)
summary_data_by_school = pd.merge(summary_draft, summary_data_by_school, on='school_name')
#add school type to the summary data, drop some columns, and re-order columns
school_types = master_df.loc[:,['school_name','type']]
summary_data_by_school = summary_data_by_school.merge(school_types, how='left', on='school_name')
summary_data_by_school = summary_data_by_school.drop_duplicates()
summary_data_by_school = summary_data_by_school.set_index('school_name')
summary_data_by_school = summary_data_by_school.drop(['School ID', 'Passing Math', 'Passing Reading', 'Passing Both'],1)
summary_data_by_school = summary_data_by_school.rename(columns={'type': 'School Type'})
summary_data_by_school = summary_data_by_school[['School Type', 'Total Students', 'Total School Budget', 'Per Student Budget',
                                                'Average Math Score','Average Reading Score', '% Passing Math', '% Passing Reading',
                                                '% Overall Passing']]

summary_data_by_school
```

MATH SCORES BY GRADE

```python
# #filter to math scores by grade
master_filtered_math = master_df.loc[:,['school_name', 'grade', 'math_score']].groupby(['school_name','grade']).mean().unstack('grade')
master_filtered_math.columns = master_filtered_math.columns.droplevel(0)
master_filtered_math = master_filtered_math[['9th', '10th', '11th', '12th']]
master_filtered_math
```

```python
#filter to reading scores by grade
master_filtered_reading = master_df.loc[:,['school_name', 'grade', 'reading_score']].groupby(['school_name', 'grade']).mean().unstack('grade')
master_filtered_reading.columns = master_filtered_reading.columns.droplevel(0)
master_filtered_reading = master_filtered_reading[['9th', '10th', '11th', '12th']]
master_filtered_reading
```

SCORES BY SCHOOL SPENDING

```python
# create bins for spending levels, group by those levels and report associated scoring statistics
bins=[0,585,630,645,660]
bin_names = ['<585', '585-630','630-645','645-660']
summary_data_by_school.loc[:,'Per Student Spending Range'] = pd.cut(summary_data_by_school.loc[:,'Per Student Budget'],bins=bins, labels=bin_names)
student_spending_bins = summary_data_by_school.loc[:,['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing','Per Student Spending Range']].groupby(['Per Student Spending Range']).mean()
student_spending_bins
```

```python
# create bins for school size, group by those levels and report associated scoring statistics
binz=[0,1000,2000,5000]
bin_namez = ['Small(<1000)', 'Medium (1000-2000)','Large (>2000)']
summary_data_by_school.loc[:,'School Size'] = pd.cut(summary_data_by_school.loc[:,'Total Students'],bins=binz, labels=bin_namez)
student_spending_binz = summary_data_by_school.loc[:,['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing','School Size']].groupby(['School Size']).mean()
student_spending_binz
```

```python
# group by school type and report associated scoring statistics
scores_by_school_type = summary_data_by_school.loc[:,['School Type', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']].groupby(['School Type']).mean()
scores_by_school_type
```
